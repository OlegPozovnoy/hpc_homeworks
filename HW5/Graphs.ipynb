```python
import matplotlib.pyplot as plt
data1 = [ 0.073597, 0.075806, 13.195486,  17.250563]
x = [1,2,3,4]
plt.plot(x, data1, marker = 'o')
plt.title("с числом нитей = 2")
plt.xlabel('Число процессов')
plt.ylabel('Время исполнения сек')
plt.show()

data2 = [0.089922, 0.051554, 0.043895,  0.023593]
plt.plot(x, data2,  marker = 'o')
plt.title("с числом нитей = 1")
plt.xlabel('Число процессов')
plt.ylabel('Время исполнения сек')
plt.show()

# по графикам очевидно сколько у меня процессоров на ноутбуке
```
